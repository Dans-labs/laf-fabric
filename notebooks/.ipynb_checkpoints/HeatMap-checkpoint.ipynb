## Heatmaps

```python
import matplotlib.pyplot as plt
import matplotlib as mpl
import numpy as np
import matplotlib.cm as cm
import numpy.random as rand
%matplotlib inline

def query():
    bound = rand.rand(2)
    print("bound={}".format(bound))
    lower_bound = bound[0] if bound[0] < bound[1] else bound[1]
    upper_bound = bound[1] if bound[0] < bound[1] else bound[0]
    lower_bound *= 100
    upper_bound *= 100
    print("From {} to {}".format(lower_bound, upper_bound))
    size = round(rand.rand(1)[0] * 10) + 10
    print("Size={}".format(size))
    data = rand.rand(size)
    data *= (upper_bound - lower_bound)
    data += lower_bound
    print("Queried data={}".format(data))
    return data

def showdata(data):
    min_value = min(data)
    print("Minvalue={}".format(min_value))
    max_value = max(data)
    relative_data = data - min_value
    print("Relative data={}".format(relative_data))
    relative_data /= (max_value - min_value)
    relative_data *= 100
    print("Relative data={}".format(relative_data))
    data_frame = np.zeros((2, len(data)))
    data_frame[0,:] = data
    data_frame[1,:] = relative_data
    plt.matshow(data_frame, cmap=cm.jet, vmin=0, vmax=100)
    for y in range(data_frame.shape[0]):
        for x in range(data_frame.shape[1]):
            plt.text(x, y, '{:.3g}'.format(data_frame[y, x]),
                 horizontalalignment='center',
                 verticalalignment='center',
            )
    plt.show()

showdata(query())
```

```python
nrows, ncols = 10,10
#image = np.zeros(nrows*ncols)
for i in range(nrows):
    image = np.zeros((1, ncols), dtype=int)
    image[0,:] = range(i+1, (ncols+1)*(i+1), i+1)
    row_labels = ["step {}".format(i+1)]
    col_labels = range(i+1, (10+1)*(i+1), i+1)
    plt.matshow(image, cmap=cm.jet)
    plt.xticks(range(ncols), col_labels)
    plt.yticks(range(1), row_labels)
    plt.show()
```
